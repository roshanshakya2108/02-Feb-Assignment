```python
#Answer 1:


# To create a Bokeh plot using Python, follow these steps:

# Install Bokeh:

# pip install bokeh

# Create a simple plot:


from bokeh.plotting import figure, show
p = figure(title="Simple Line Example", x_axis_label='x', y_axis_label='y')
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], legend_label="Temp.", line_width=2)
show(p)
```

```python
#Answer 2:
# Glyphs are the basic visual building blocks in Bokeh. They are shapes like circles, squares, lines, etc., that can be added to a plot to represent data points.

from bokeh.plotting import figure, show

# Create a new plot
p = figure(title="Glyph Example", x_axis_label='x', y_axis_label='y')

# Add circle glyphs
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=10, color="navy", alpha=0.5)

# Show the results
show(p)
```

```python
#Answer 3:

from bokeh.plotting import figure, show

# Create a new plot
p = figure(title="Customized Plot", x_axis_label='x', y_axis_label='y')

# Customize title
p.title.text = "My Customized Plot"
p.title.align = "center"
p.title.text_font_size = "20px"
p.title.text_color = "orange"

# Customize axes
p.xaxis.axis_label = "X-Axis"
p.xaxis.axis_label_text_color = "blue"
p.yaxis.axis_label = "Y-Axis"
p.yaxis.axis_label_text_color = "green"

# Add line glyph with customized legend
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], legend_label="Temperature", line_width=2)

# Customize legend
p.legend.title = "Legend"
p.legend.title_text_font_size = "15px"
p.legend.label_text_font_size = "10px"

# Show the results
show(p)
```

```python
# prompt: What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

# A Bokeh server is a Python application that runs a Bokeh document and exposes it to web browsers.
# This allows users to interact with the plot and see updates in real time.

# To use a Bokeh server, you can follow these steps:

# 1. Install Bokeh:

# pip install bokeh

# 2. Create a Bokeh document:

# from bokeh.plotting import figure, show

# p = figure(title="Interactive Plot")
# p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

# 3. Run the Bokeh server:

# bokeh serve --show my_document.py

# 4. Open a web browser and navigate to the URL provided by the server.

# 5. Interact with the plot to see updates in real time.

# You can also use the Bokeh server to create more complex interactive applications, such as dashboards and data visualizations.
```

```python
#Answer 5:

# Import necessary libraries
from bokeh.embed import components
from flask import Flask, render_template

# Create a Flask app
app = Flask(__name__)

# Define a route for the web page
@app.route('/')
def index():
    # Create a Bokeh plot
    p = figure(title="Embedded Bokeh Plot", x_axis_label='x', y_axis_label='y')
    p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

    # Generate the necessary components for embedding the plot
    script, div = components(p)

    # Render the web page with the embedded plot
    return render_template('index.html', script=script, div=div)

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=True)
```
