Scatter Plot (sns.scatterplot):

Use: Visualizes the relationship between two numerical variables. It helps identify correlations, trends, and outliers.
Line Plot (sns.lineplot):

Use: Displays data points connected by straight line segments. It is useful for visualizing trends over time or ordered categories.
Box Plot (sns.boxplot):

Use: Summarizes the distribution of a dataset by showing the median, quartiles, and potential outliers. It is useful for comparing distributions across multiple groups.
Histogram (sns.histplot):

Use: Visualizes the distribution of a single numerical variable by dividing the data into bins. It is helpful for understanding the frequency distribution and identifying the shape of the data.
Heatmap (sns.heatmap):

Use: Displays data in a matrix format using color to indicate values. It is useful for visualizing the intensity of values across a 2D space, such as correlation matrices or frequency tables.


```python
#Answer  2:

import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot a line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title('FMRI Signal over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()

```

```python
#Answer 3 :

titanic = sns.load_dataset("titanic")

# Plot box plot for age
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Plot box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
#Answer 4 :

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot a histogram for price with hue for cut
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
#Answer 5 :

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot a pair plot with hue for species
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```
