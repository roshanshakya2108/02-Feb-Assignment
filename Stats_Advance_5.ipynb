```python
#Answer  1:

import scipy.stats as stats

mean = 50
std_dev = 5
sample_size = 30  # Assuming a sample size

# Calculate the standard error
std_err = std_dev / (sample_size ** 0.5)

# Z-score for 95% confidence
z_score = stats.norm.ppf(0.975)

# Calculate the confidence interval
margin_of_error = z_score * std_err
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
confidence_interval
```

```python
#Answer  2:

import numpy as np
import scipy.stats as stats

# Observed frequencies (replace these with the actual observed counts)
observed_frequencies = [20, 30, 25, 10, 10, 5]

# Expected frequencies based on given proportions
expected_proportions = [0.20, 0.20, 0.20, 0.10, 0.10, 0.20]
total_count = sum(observed_frequencies)
expected_frequencies = [total_count * prop for prop in expected_proportions]

# Conduct the chi-square goodness of fit test
chi_square_stat, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)
chi_square_stat, p_value
```

```python
#Answer  4:

import scipy.stats as stats

n = 500
x = 60
p_hat = x / n

# Z-score for 95% confidence
z_score = stats.norm.ppf(0.975)

# Calculate the standard error
std_err = (p_hat * (1 - p_hat) / n) ** 0.5

# Calculate the confidence interval
margin_of_error = z_score * std_err
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)
confidence_interval
```

```python
#Answer  6:

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df = 10
chi_square_stat = 15

# Generate the chi-square distribution
x = np.linspace(0, 30, 500)
y = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, y, label='Chi-square Distribution')
plt.fill_between(x, y, where=(x <= chi_square_stat), color='skyblue', alpha=0.5)
plt.axvline(chi_square_stat, color='red', linestyle='--', label=f'Chi-square stat = {chi_square_stat}')
plt.title('Chi-square Distribution with 10 degrees of freedom')
plt.xlabel('Chi-square value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
#Answer 11:

import scipy.stats as stats

sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean = 70

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

# Degrees of freedom
df = sample_size - 1

# Calculate the p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))
t_statistic, p_value
```
