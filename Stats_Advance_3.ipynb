Estimation Statistics is a branch of statistics that focuses on estimating the population parameters based on sample data. It involves two main types of estimates:

1. Point Estimate: A single value that serves as an estimate of a population parameter.

2. Interval Estimate: A range of values, derived from the sample data, that is likely to contain the population parameter. An example of this is the confidence interval for the population mean, which provides an estimated range that is likely to include the true population mean


```python
#Answer 2  :

from scipy.stats import t
import numpy as np

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level=0.95):
    alpha = 1 - confidence_level
    df = sample_size - 1
    t_critical = t.ppf(1 - alpha/2, df)
    
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
    
    return confidence_interval

# Example usage
sample_mean = 500
sample_std = 50
sample_size = 50
confidence_level = 0.95

estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level)

```

```python
#Answer 5  :

from scipy.stats import ttest_ind

def hypothesis_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)
    result = "Fail to reject the null hypothesis" if p_value > alpha else "Reject the null hypothesis"
    
    return t_stat, p_value, result

# Example usage
sample1 = [80, 85, 88, 90, 95]  # Sample data for population 1
sample2 = [78, 82, 85, 87, 89]  # Sample data for population 2

alpha = 0.05
t_stat, p_value, result = hypothesis_test(sample1, sample2, alpha)
print(f"T-statistic: {t_stat}, P-value: {p_value}, Result: {result}")

```

```python
#Answer  13 :

from scipy.stats import t
import numpy as np

sample_mean = 500
sample_std = 50
sample_size = 50
confidence_level = 0.95

# Calculate the t-critical value
alpha = 1 - confidence_level
df = sample_size - 1
t_critical = t.ppf(1 - alpha/2, df)

# Calculate the margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
confidence_interval
```

```python
#Answer  16 :

n1 = 30
mean1 = 80
std1 = 10

n2 = 40
mean2 = 75
std2 = 8
alpha = 0.01

# Calculate the t-statistic
pooled_std = np.sqrt((std1**2/n1) + (std2**2/n2))
t_statistic = (mean1 - mean2) / pooled_std

# Calculate the degrees of freedom
df = min(n1-1, n2-1)

# Calculate the critical t-value
t_critical = t.ppf(1 - alpha/2, df)

# Decision
result = "Reject the null hypothesis" if abs(t_statistic) > t_critical else "Fail to reject the null hypothesis"
t_statistic, t_critical, result

```
