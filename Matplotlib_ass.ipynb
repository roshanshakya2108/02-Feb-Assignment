Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations. It is widely used in data analysis and scientific research for its ability to produce high-quality graphs and plots.

Why it is used:

To visualize data in a clear and understandable manner.
To create a variety of plots and charts for exploratory data analysis.
To customize the appearance of plots to suit specific requirements.
To save plots in various formats like PNG, PDF, SVG, etc.
Five plots that can be plotted using the Pyplot module of Matplotlib:

Line plot
Scatter plot
Bar plot
Histogram
Box plot

```python
#Answer 2 :
# Scatter plot is a type of plot that shows the relationship between two numerical variables. 
# Each point on the plot represents an observation with its x and y coordinates indicating values for the two variables.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot of Random Data')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()

```
